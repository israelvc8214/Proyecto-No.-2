### **Problema 1:** 
#### Generar una función y utilizar esta para recodificar la variable “país de origen” (1, 2, 3) en el dataframe a través de la función apply.

```python
# Importamos los paquetes o librerias que vamos a utilizar
import pandas as pd
import numpy as np
```

```python
# Para utilizar el archivo autos.csv, en un caso personal, yo lo subí a GitHub por cuestiones de practicidad. En este sentido,
# vamos a generar el primer dataframe a partir del archivo almacenado en el repositorio mencionado. Como el archivo viene separado
# por punto y coma, vamos a ingresar el separador "sep", seguido por el punto y coma. 
csv_path = ('https://raw.githubusercontent.com/israelvc8214/Proyecto-No.-2/main/autos.csv')
df = pd.read_csv(csv_path, sep=';')
df
```

```python
# Comprobamos si el archivo tiene datos perdido mediante el metodo df.info y df.isnull, por lo que podemos observar que 
# sí existe 14 observaciones nulas en las variables Millas_Por_Galon y Caballo_De_Fuerza
df.info()
df.isnull().sum()
```

```python
# Imputamos los valores perdidos mediante la media para las variables que presentan esta caracteristica de missing values, por lo
# que aplicamos la siguiente instrucción al dataframe: 
df['Millas_Por_Galon'].fillna(df['Millas_Por_Galon'].mean(), inplace=True)
df['Caballo_De_Fuerza'].fillna(df['Caballo_De_Fuerza'].mean(), inplace=True)
df
```

```python
# Nuevamente comprobamos si se eliminaron o suplieron los missing values en las variables mencionadas
# y podemos observar que ya no existen datos nulos
df.info()
df.isnull().sum()
```

```python
# Una vez comprobado que ya no existen valores perdidos en el dataframe, pasamos a recodificar la variable Pais_Origen con las categorias
# 1, 2 y 3. Para el caso de la etiqueta USA se le asigna el digito 1, a Europe el digito 2, y a Japan el digito 3, esto mediante la siguiente
# funcion
def origen_country(x): #Generamos una etiqueta
  """
  Esta función reetiqueta la variable Pais_Origen
  
  """
  if x=='USA': return 1
  if x=='Europe': return 2
  if x=='Japan': return 3
df['origen_pais'] = df['Pais_Origen'].apply(origen_country)
df
```

## **Problema 2:** 
#### Generar un Notebook Colab con código y resultados, que contiene información de autos de un periodo de tiempo. Resolver los siguientes: 
##### -Limpieza y tratamiento de las variables (Según corresponda)
##### -Generar 5 nuevas variables aplicando los siguientes criterios con las variables numéricas:
##### 1. Log10(var1/var2)
##### 2. Sqrt(var1)*exp(var2)/200
##### 3. Si var 1 > var 2 → 5, caso contrario 3
##### 4. 1/logn(var1/var2)*100
##### 5. var2**2/var1



```python
# Para el segundo ejercicio, generamos una copia del primer dataframe en un nuevo contendor
df1=df
df1
```

```python
# Para generar la nueva variable millas_cil a partir del criterio Log10(var1/var2) con el logaritmo base 10, ya se hizo la limpieza de los
# datos a partir del primer dataframe y la imputación respectiva, por lo que se procede a generar la nueva variable con el criterio respectivo
# utilizando como var1=Millas_Por_Galon y var2=Cilindros, aplicando la siguiente función: 
df1['millas_cil']= np.log10(df1['Millas_Por_Galon']/df1['Cilindros'])
df1
# Para tal efecto, generamos la variable millas_cil y obtenemos el logaritmo base 10 mediante el metodo np.log10 de numpy
```

```python
# Para el caso del segundo criterio Sqrt(var1)*exp(var2)/200 se sigue utilizando las mismas variables y mediante numpy se calcula la raiz
# cuadrada para la variable 1 Millas_Por_Galon y el exponencial para la variable 2 Cilindros, esto mediante la siguiente función: 
df1['millas_cil1']= np.sqrt(df1['Millas_Por_Galon'])*np.exp(df1['Cilindros'])/200
df1
# Se genera el dataframe con la variable millas_cil1
```

```python
# Para el criterio (var 1 > var 2 → 5, caso contrario 3) se vuelve a utilizar una función similar a la del primer reetiquetado con las mismas
# variables, por lo que definimos milla_cil y utilizamos las condiciones if-else para que nos retorne 5 si Millas_Por_Galon' > 'Cilindros, o
# en caso contrario no retorne 3, generando la variable millas_cil2 mediante el metodo apply como en el primer ejercicio. 
def milla_cil(x):
  """
  Reetiqueta una variable segun ciertas condiciones
  
  """
  if  'Millas_Por_Galon' > 'Cilindros':
    return 5
  else:
    return 3

df1['millas_cil2'] = df1.apply(milla_cil, axis=1)
df1
```

```python
# Para el caso del criterio 1/logn(var1/var2)*100 utilizamos numpy para obtener un logaritmo base n, para este caso yo elijo el logaritmo base 2
# como n de acuerdo con mi perspectiva, por lo que se genera la siguiente función que crea la variable millas_cil3: 
df1['millas_cil3']= 1/np.log2(df1['Millas_Por_Galon']/df1['Cilindros'])*100
df1
```

```python
# Finalmente, el criterio var2**2/var1 los generamos con la siguiente función para crear la variable millas_cil4: 
df1['millas_cil4']= (df1['Millas_Por_Galon']**2)/(df1['Cilindros'])
df1
```

```python
# Nuevamente corroboramos que no existan valores perdidos para continuar con el ejercicio 3: 
df1.info()
df1.isnull().sum()
# Observamos que las variables generadas no contienen valores perdidos. 
```

## **Problema 3:** 
#### Crear una función que permita generar 3 variables cualitativas ordinales a partir de las variables numéricas con los siguientes criterios:
#### ★ Si es mayor a la “media + 1*desviación estándar” → Alto
#### ★ Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
#### ★ Si es menor a la “media - 1*desviación estándar” → Bajo
#### Nota: Mostrar con las nuevas variables generadas. Elegir con libertad las variables a trabajar.
#### La función debe contar con su respectiva documentación.

```python
# Para este primer caso, lo primero que hago es generar la media para la variable
# Millas_Por_Galon a partir de la cual vamos a generar la primer variable cualitativa ordinal
media = df1.Millas_Por_Galon.mean()
media
```

```python
# Generamos la desviación estandar importando la libreria statistics
import statistics 
desvest = statistics.stdev(df1.Millas_Por_Galon)
desvest
```

```python
# Generamos nuevamente la desviación estandar con otro metodo y nos da el mismo resultado que la anterior
desve = df1.Millas_Por_Galon.std()
desve
```

```python
# Generamos media + 1*desviación estándar de Millas_Por_Galon
crit1=media+1*desvest
crit1
```

```python
# Generamos media - 1*desviación estándar para Millas_Por_Galon con el fin de poder aplicar los 3 criterios solicitados. 
crit2=media-1*desvest
crit2
```

```python
# Una vez generados los criterios, en mi caso, yo utilizo los resultados de crit1 y crit2 para aplicar los 3 criterios, definiendo
# millas_galon y aplicando los 3 criterios, para posteriormente definir la variable primera variable cualitativa millas_galon1
def millas_galon(x):
  """"
  Devuelve una variable cualitativa con criterios Alto, Medio y Bajo

  """
  if x > 31.252977248514842: return 'Alto' # Si es mayor a la “media + 1*desviación estándar” → Alto
  if 15.776168480128373 < x < 31.252977248514842: return 'Medio' # Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
  if x < 15.776168480128373: return 'Bajo' # Si es menor a la “media - 1*desviación estándar” → Bajo
df1['millas_galon1'] = df1['Millas_Por_Galon'].apply(millas_galon)
df1
```

```python
# Generamos la desviación estandar con otro metodo
desve1 = df1.Peso_En_Libras.std()
desve1
```

```python
# Generamos media + 1*desviación estándar de Peso_En_Libras
crit11=media1+1*desvest1
crit11
```

```python
# Una vez generados los criterios, en mi caso, yo utilizo los resultados de crit11 y crit22 para aplicar los 3 criterios, definiendo
# peso_libras y aplicando los 3 criterios, para posteriormente definir la variable segunda variable cualitativa peso_libra1
def peso_libras(x):
  """"
  Devuelve una variable cualitativa con criterios Alto, Medio y Bajo

  """
  if x > 3826.418121342799: return 'Alto' # Si es mayor a la “media + 1*desviación estándar” → Alto
  if 2132.4094648640976 < x < 3826.418121342799: return 'Medio' # Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
  if x < 2132.4094648640976: return 'Bajo' # Si es menor a la “media - 1*desviación estándar” → Bajo
df1['peso_libra1'] = df1['Peso_En_Libras'].apply(peso_libras)
df1
```

```python
# Una vez generados los criterios, en mi caso, yo utilizo los resultados de crit111 y crit222 para aplicar los 3 criterios, definiendo
# desplazamiento1 y aplicando los 3 criterios, para posteriormente definir la variable segunda variable cualitativa desplazamiento2
def desplazamiento1(x):
  """"
  Devuelve una variable cualitativa con criterios Alto, Medio y Bajo

  """
  if x > 299.70201502973504: return 'Alto' # Si es mayor a la “media + 1*desviación estándar” → Alto
  if 89.85709827075755 < x < 299.70201502973504: return 'Medio' # Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
  if x < 89.85709827075755: return 'Bajo' # Si es menor a la “media - 1*desviación estándar” → Bajo
df1['desplazamiento2'] = df1['Desplazamiento'].apply(desplazamiento1)
df1
```

## **Problema 5:** 
#### Crear una función que devuelva un mapa de calor con la correlación de todas las variables numéricas (incluye las generadas), para cada país.

```python
# Nuevamente importamos matplotlib y seaborn para los gráficos correspondientes
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correlación de todas las variables (Cualitativas y Cuantitativas) respecto a la variable pais
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df1.corr()[['origen_pais']].sort_values(by='origen_pais', ascending=False), annot=True, cmap='Blues')
heatmap.set_title('Correlating with País de Origen', fontdict={'fontsize':18}, pad=16);
```

```python
# Correlacion general de variables cuantitativas sin tomar en cuenta a cada país
# Generamos una lista con las variables cuantitativas generadas en el dataframe df1 y a su vez generamos un contenedor df2 con las variables
# númericas elegidas
columnas=['Millas_Por_Galon', 'Cilindros',	'Peso_En_Libras', 'Desplazamiento',	'Aceleracion', 'Caballo_De_Fuerza', 'millas_cil',	'millas_cil1', 'millas_cil3', 'millas_cil4']
df2=df1[columnas]
df2
```

```python
# Generamos una correlación general sin filtrar los paises a partir de las variables contenidas en df2
plt.figure(figsize=(14, 10))
correlacion = df2.corr() 

sns.heatmap(correlacion,
            xticklabels = correlacion.columns.values,
            yticklabels = correlacion.columns.values,
            annot = True,
            cmap = "Blues" 
            );
```

```python
# Generamos el mapa de correlación de Pearson para USA utilizando el contenedor df3, obteniendo las siguientes correlaciones:
plt.figure(figsize=(14, 10))
correlacion_usa = df3.corr() 

sns.heatmap(correlacion_usa,
            xticklabels = correlacion_usa.columns.values,
            yticklabels = correlacion_usa.columns.values,
            annot = True,
            cmap = "Blues" 
            );
```
